# Dynamics evolution

```python
import numpy as np
import glob
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style as style 
style.available
sns.set_context('paper')
sns.set()
SaveFiles="DiagElement"
dataset="ENZYMES"
modelName="ChebConvNet"
NumLayers=2
FileConstrant2="param"
num_epochs=300
interal=40
for epoch in range(num_epochs):
    if epoch%interal==0 or epoch==num_epochs-1:
        save_png_name="{}-{}-{}-{}-{}.png".format(SaveFiles,dataset,modelName,NumLayers,epoch)
        EvolutionDynamics=np.load("{}Convergence/DiagElement-{}-{}-{}_1024_{}_0.99_0.1-monte_0-{}.npy".format(dataset,dataset,modelName,FileConstrant2,NumLayers,epoch),allow_pickle=True)
        print(EvolutionDynamics.shape)
        for i in range(EvolutionDynamics.shape[1]):
            plt.plot(np.linspace(1,NumLayers,NumLayers),np.array(EvolutionDynamics.T)[i])
        plt.legend([1,2,3,4,5])
        plt.xlabel("Layer")
        plt.ylabel("")
        plt.savefig(save_png_name,dpi=600)
        plt.show()
```
